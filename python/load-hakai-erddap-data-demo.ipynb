```python
import pandas as pd
```

```python
# Let's download the last two montsh of data from QU39 in a csv format
url = 'https://catalogue.hakai.org/erddap/tabledap/HakaiWaterPropertiesInstrumentProfileProvisional.csv?&station="QU39"&time>now-2months';

# Load data and ignore the second row which is the units associated with each columns in ERDDAP csv format
df = pd.read_csv(url, skiprows=[1])
df.head()
```

```python
# Convert time strings to datetime objects
df['time'] = pd.to_datetime(df['time'])
```

```python
fig = df.plot.scatter(x='time',y='depth',c='temperature')
fig.axes.invert_yaxis()
```
